# Starter Code

```python
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import Lasso
```

```python
# Load Fashion MNIST dataset
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()
```

Each row is an item of clothing.

Each column or feature is a pixel value.

# Preprocessing

```python
# Flatten the images from 28x28 to 784 features (28*28)
X_train = X_train.reshape(X_train.shape[0], -1)
X_test = X_test.reshape(X_test.shape[0], -1)
```

# Baseline

## RBF Kernel

```python
# Initialize an SVM classifier
svm_classifier = SVC(kernel='rbf')

# Train the model in batches of 5000
batch_size = 5000
n_batches = X_train.shape[0] // batch_size

for i in range(n_batches):
    start = i * batch_size
    end = (i + 1) * batch_size
    print(f"Training batch {i+1}/{n_batches}")

    # Extract the current batch
    X_batch = X_train[start:end]
    y_batch = y_train[start:end]

    # Fit the model on the current batch
    svm_classifier.fit(X_batch, y_batch)

# Make predictions on the test data
y_pred = svm_classifier.predict(X_test)

# Evaluate the accuracy of the classifier
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy * 100:.2f}%")
```

### Confusion Matrix

```python
# Generate and display the confusion matrix
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=np.unique(y_test))
disp.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.show()
```

# KBest and SVM

```python
# Define batch size
batch_size = 5000

# Use a batch of 5000 samples for both data and target
X_batch = X_train[:batch_size]
y_batch = y_train[:batch_size]

# Feature selection using SelectKBest (k=300)
k_best = SelectKBest(score_func=chi2, k=300)
X_train_kbest = k_best.fit_transform(X_batch, y_batch)  # Use X_batch and y_batch for fitting
X_test_kbest = k_best.transform(X_test)

# Train an SVM model with an RBF kernel using the selected features
svm_rbf = SVC(kernel='rbf')
svm_rbf.fit(X_train_kbest, y_batch)  # Fit the model using the batch of labels

# Make predictions on the test data
y_pred = svm_rbf.predict(X_test_kbest)

# Evaluate the model's accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy with KBest (k=300): {accuracy * 100:.2f}%")
```

This model is worse off than the baseline, since the baseline has 83.28% accuracy while this model has 81.90%.

## Confusion Matrix

```python
# Generate and display the confusion matrix
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=np.unique(y_test))
disp.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.show()
```

## Classification Report

```python
# Classification Report (Precision, Recall, F1-score)
class_report = classification_report(y_test, y_pred, target_names=[
    'T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
    'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot'])
print(class_report)
```

```python
# Visualizing Precision, Recall, and F1-score
report = classification_report(y_test, y_pred, output_dict=True)
scores_df = pd.DataFrame(report).T[:-3]  # Exclude avg metrics
scores_df[['precision', 'recall', 'f1-score']].plot(kind='bar')
plt.title('Precision, Recall, and F1-Score by Class')
plt.xticks(rotation=0)
plt.show()
```

# DTC and SVM

```python
# Define batch size
batch_size = 5000

# Create a batch of data and corresponding labels
X_batch = X_train[:batch_size]
y_batch = y_train[:batch_size]

# Train a Decision Tree Classifier to get feature importances
decision_tree = DecisionTreeClassifier(random_state=42)
decision_tree.fit(X_batch, y_batch)  # Use y_batch instead of y_train

# Rank features by importance and select the top 100
importances = decision_tree.feature_importances_
indices = np.argsort(importances)[::-1]  # Sort in descending order

# Select the top 100 features
top_100_indices = indices[:100]
X_train_top100 = X_batch[:, top_100_indices]
X_test_top100 = X_test[:, top_100_indices]

# Train an SVM model using the top 100 features
svm_rbf = SVC(kernel='rbf')
svm_rbf.fit(X_train_top100, y_batch)

# Make predictions and evaluate the model
y_pred = svm_rbf.predict(X_test_top100)
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy with Top 100 Features: {accuracy * 100:.2f}%")
```

This model is also worse off than the baseline since it has 81.96% accuracy which is lower than the baseline's accuracy.

# LASSO and SVM

```python
# Define batch size
batch_size = 5000

# Create a batch of training data
X_batch = X_train[:batch_size]
y_batch = y_train[:batch_size]

# Train a Lasso model for feature selection with alpha=0.05
lasso = Lasso(alpha=0.05)
lasso.fit(X_batch, y_batch)  # Use X_batch and corresponding y_batch

# Select features with non-zero coefficients
selected_features = np.where(lasso.coef_ != 0)[0]
print(f"Selected {len(selected_features)} features out of {X_batch.shape[1]}")

# Apply the selected features to both the train and test sets
X_train_lasso = X_batch[:, selected_features]  # Apply the features on the batch data only
X_test_lasso = X_test[:, selected_features]    # Apply the same features on the test set

# Train an SVM model using the selected features
svm_rbf = SVC(kernel='rbf')
svm_rbf.fit(X_train_lasso, y_batch)  # Train on the batch data

# Make predictions and evaluate the model
y_pred = svm_rbf.predict(X_test_lasso)
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy with Lasso Feature Selection: {accuracy * 100:.2f}%")
```

This model is better than the baseline as the baseline has 83.28% accuracy while this model has 83.41%. It is only better by a very low margin.

# PCA and SVM

```python
# Define batch size
batch_size = 5000

# Create a batch of training data and corresponding labels
X_batch = X_train[:batch_size]
y_batch = y_train[:batch_size]

# Apply PCA for dimensionality reduction (retain 95% variance)
pca = PCA(n_components=0.95)
data_train_pca = pca.fit_transform(X_batch)  # Apply PCA on the batch of training data
data_test_pca = pca.transform(X_test)        # Apply PCA on the entire test data

# Train an SVM model with an RBF kernel using the transformed data
svm_rbf = SVC(kernel='rbf')
svm_rbf.fit(data_train_pca, y_batch)  # Use the batch data and batch labels

# Make predictions on the test data
y_pred = svm_rbf.predict(data_test_pca)

# Evaluate the model's accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy with PCA: {accuracy * 100:.2f}%")
```

This model is better than the baseline as the baseline has 83.28% accuracy while this model has 84.03%.
